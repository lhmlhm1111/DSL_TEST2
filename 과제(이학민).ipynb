10) `kobis` 데이터의 `genre` 컬럼은 영화의 장르를 나타내며, 영화가 복수의 장르에 해당할 경우 컴마 `,` 로 구분하였습니다. 이 컬럼을 더미 인코딩하되, 빈도가 100 이상인 장르들만 남겨주세요. 이후 원래의 `genre` 컬럼은 드랍해주세요.

```python
uniqueVals = np.unique(np.array(",".join(kobis.genre.values).split(",")))
uniqueVals
```

```python
def dummyEncoding(uniqueVal):
    dummy = kobis.genre.apply(lambda x: uniqueVal in x).astype(int).rename(uniqueVal)
    if dummy.sum() > 100:
        return dummy
```

```python
dummyData = pd.concat(list(map(dummyEncoding, uniqueVals)), axis=1)
pd.concat([kobis, dummyData], axis=1).drop("genre", axis=1) #이거 못풀었다 ㅠ 어렵누
```
